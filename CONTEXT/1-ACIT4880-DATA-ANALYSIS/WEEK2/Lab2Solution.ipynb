````python
weight_lb = np.array([125, 98, 135, 170, 162, 147, 144, 177, 180, 112])
np_weight_kg = weight_lb * 0.453592

np_height_m = np.array([1.77, 1.45, 1.85, 2.2, 1.75, 1.66, 1.70, 1.89, 2.1, 1.6])

BMI = np_weight_kg / np_height_m ** 2
print(BMI)

light = BMI < 21
print(light)

print(BMI[light])
print(np.mean(np_height_m))
print(np.median(np_height_m))
````
