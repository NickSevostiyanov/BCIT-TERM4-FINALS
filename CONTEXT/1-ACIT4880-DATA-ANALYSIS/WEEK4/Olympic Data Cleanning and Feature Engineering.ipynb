### Targets

```python
df = bios.copy()
```

```python
len(df[df['Used name'].str.contains("\^")])
```

```python
df["Clean name"] = df['Used name'].str.replace("•|\^", " ", regex=True)
```

```python
df.sample(5)
```

```python
dummy_data = pd.DataFrame(['180 cm', '180 cm / 68 kg', '78 cm / 74 kg', '50 kg'], columns=["Measurements"])
dummy_data
```

```python
dummy_data[ ["Height_cm", "Weight_kg"] ] = dummy_data["Measurements"].str.split('/', expand=True)
dummy_data
```

```python
df[["Height_cm", "Weight_kg"]] = df["Measurements"].str.split('/', expand=True)
df.sample(5)
```

```python
df["Height_cm"] = df["Height_cm"].str.extract(r'(\d+)')
df["Weight_kg"] = df["Weight_kg"].str.extract(r'(\d+)')
df.sample(5)
```

```python
Date_pattern = r'(\d+[\s|-]\w+[\s|-]\d+)'
df["Born_date"] = df["Born"].str.extract(Date_pattern)
```

```python
df.sample(5)
```

```python
df["Died_date"] = df["Died"].str.extract(Date_pattern)
```

```python
df.sample(5)
```

```python
Location_Pattern = r'in\s([\w\s()-]+),\s([\w\s()-]+)\s\((\w+)\)'
df[ ["Born_city", "Born_region", "Born_country"] ]  = df["Born"].str.extract(Location_Pattern, expand=True)
df.sample(5)
```

```python
columns_to_keep = ["Clean name", "Sex", "Height_cm", "Weight_kg", "Born_date", "Died_date", "Born_city", "Born_region", "Born_country"]
df_clean=df[columns_to_keep]
df_clean.sample(5)
```
