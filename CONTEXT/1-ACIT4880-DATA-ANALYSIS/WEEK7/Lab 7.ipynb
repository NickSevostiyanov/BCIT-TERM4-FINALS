<h3> Example 1. Find the 95% confidence t-interval for the mean number of Customer Service
Calls (Use the churn data set)
    - slide 13 in chapter4

```python
#Ex1 solution here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from math import *


df = pd.read_csv("churn.csv")
cust_servcalls = df['CustServ Calls']
cust_servcalls_mean = cust_servcalls.mean()
cust_servcalls_std = cust_servcalls.std()
cust_servcalls_count = len(cust_servcalls)
t_value = stats.t.ppf(0.975, 3332)
conf_interval_from = cust_servcalls_mean - t_value * (cust_servcalls_std / sqrt(cust_servcalls_count))
conf_interval_to = cust_servcalls_mean + t_value * (cust_servcalls_std / sqrt(cust_servcalls_count))
print (conf_interval_from, conf_interval_to)
```

<h3> Example 2. Calculate 95% confidence interval for the proportion π of churners among the
entire population of the company’s customers (Use churn data set) - slide 26 in Chap4

```python
#Ex2 solution here
from scipy import stats
from math import *

cust_churn = df['Churn?']
cust_churn_count = len(cust_churn)
cust_churn_rate = len(cust_churn[cust_churn == 'True.']) / len(cust_churn)

z_value = stats.norm.ppf(0.975)


conf_interval_from = cust_churn_rate - z_value * sqrt((cust_churn_rate * (1 - cust_churn_rate))/cust_churn_count)
conf_interval_to = cust_churn_rate + z_value * sqrt((cust_churn_rate * (1 - cust_churn_rate))/cust_churn_count)
print (conf_interval_from, conf_interval_to)
```

### Example 3: Test whether mean number of Customer Service Calls for customers
with International Plan and VoiceMail Plan and who have more than 220 day
minutes differs from 2.4, with a level of significance α=0.05

- 𝐻0: 𝜇 = 2.4 𝑣s 𝐻𝑎: 𝜇 ≠ 2.4
- Null hypothesis is rejected if p-value<0.05
- We also know that:
   - µ0 = 2.4 𝑥̅= 1.607
   - 𝑠 = 1.892 𝑛 = 28

```python
tt = (1.607 - 2.4)/ (1.892/sqrt(28))
print(tt)

pval = stats.t.sf(np.abs(tt), 27)*2
print(pval)
```
